## Load the data in the folder `data/building_consumption` into a pandas dataframe.

Data is separated in 4 files. Don't forget to concatenate them, converte the date column to a datetime object and set it as index, and sort the index.

```python
import os
# get files in the folder
files = os.listdir('data/building_consumption/')
print('located files', files)
```

```python
# load the data
import pandas as pd

df = pd.DataFrame()
for file in files:
    print('loading file', file)
    df_temp = pd.read_csv('data/building_consumption/' + file, header=0, sep=',',  parse_dates=True)
    print('loaded file', file, 'with shape', df_temp.shape)
    df = pd.concat([df, df_temp], axis=0)
print('loaded file', file, 'with shape', df.shape)

df.head()
```

```python
# convert the date column to datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%d/%m/%Y %H:%M:%S')

# set the date column as index
df.set_index('Timestamp', inplace=True)

# sort the index
df.sort_index(inplace=True)

df.head()
```

## get the columns names for each type


```python
corrente_columns = [c for c in df.columns if 'Corrente' in c]
tensao_columns = [c for c in df.columns if 'Tensao' in c]
cost_columns = [c for c in df.columns if (('COST' in c) or ('Cost' in c))]

print('corrente_columns', corrente_columns)
print('tensao_columns', tensao_columns)
print('cost_columns', cost_columns)
```

```python
df.plot(figsize=(15, 5))
```

## Check if there are any missing values. If there are, fill them with the previous value.

```python
print(df.isna().sum())

df.ffill(inplace=True)

print(df.isna().sum())
```

## Group the data by day of week and plot the mean consumption for each day.
Can you comment on the results?

```python
df.groupby(df.index.dayofweek).mean().plot(figsize=(15, 5))
```

## resample the data by day and plot the min, mean and maximum consumption for each day.

```python
ax = df.resample('D').min().plot(figsize=(15, 5))
df.resample('D').max().plot(ax=ax)
df.resample('D').mean().plot(ax=ax)

ax.legend(['min', 'max', 'mean'])
```

## Decompose the data into trend, seasonality and residuals.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(df, model='additive', period=24)

_ = decomposition.plot()
```

## Resample the data for hourly values. Plot the data for the months of February, March and April 2023.

Can you spot the weekends? and the Carnaval holidays?

```python
df_hourly = df.resample('H').mean().dropna()

for m in [2, 3, 4]:
    mask = (df_hourly.index.month == m) & (df_hourly.index.year == 2023)
    df_hourly[mask].plot(figsize=(15, 5))
```

## Do the decomposition for the months of of February, March and April 2023.

```python
for m in [2, 3, 4]:
    mask = (df_hourly.index.month == m) & (df_hourly.index.year == 2023)
    decomposition = seasonal_decompose(df_hourly[mask], model='additive', period=24)
    _ = decomposition.plot()
```
